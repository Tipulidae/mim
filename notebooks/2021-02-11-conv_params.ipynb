# Analyzing filters

```python
temp = s.kernel.astype(str) + s.pool.astype(str)
temp.name = 'temp'
num_filters = []
for _, df in s.join(temp).groupby('temp'):
    if len(filters := df.drop_duplicates(subset=['filters']).sort_values(by='filters').index.values) > 3:
        print(filters)
        num_filters.append(filters)
```

```python
k = 0
display(s.loc[num_filters[k], :])
p.plot_history(num_filters[k], columns=['val_loss'], figsize=(15, 8), grid=True)
```

```python
k = 1
display(s.loc[num_filters[k], :])
p.plot_history(num_filters[k], columns=['val_loss'], figsize=(15, 8), grid=True)
```

```python
k = 2
display(s.loc[num_filters[k], :])
p.plot_history(num_filters[k], columns=['val_loss'], figsize=(15, 8), grid=True)
```

# Analyzing kernel size

```python
temp = s.filters.astype(str) + s.pool.astype(str)
temp.name = 'temp'
kernels = []
for _, df in s.join(temp).groupby('temp'):
    if len(kernel := df.drop_duplicates(subset=['kernel']).sort_values(by='kernel').index.values) > 3:
        print(kernel)
        kernels.append(kernel)
```
